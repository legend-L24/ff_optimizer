```python

def custom_uppercase(s):
    elements = ['Al', 'Mg', 'Ca']  # 
    result = []
    i = 0    
    while i < len(s):
        if i + 1 < len(s) and s[i:i+2] in elements:  # 
            result.append(s[i:i+2])  # 
            i += 2
        else:
            result.append(s[i].upper())  # 
            i += 1    
    return ''.join(result)# 
test_string = 'hello Al and Mg, Ca is great!'
result = custom_uppercase(test_string)
print(result)  # : 'HELLO Al and Mg, Ca is great!'
```

# Plot the binding energies from DFT and different kinds of force filed.

```python
from aiidamof import AiidaMofs, merge_aiida_mofs_list, custom_uppercase

element_lst = ['Al','Ga','In']
Isotherm_unit = "mol/kg" # this term does not matter in ploting binding energy

mofs_ls =  [[],[]]

for idx, forcefield in enumerate([ "Al_graphite","UFF"]):
    for element in element_lst:
        path = "_".join([forcefield, element, "paper.log"])
        aiidamofs = AiidaMofs(path, Isotherm_unit)
        for aiidamof in aiidamofs.mofs:
            aiidamof.mofname = custom_uppercase(aiidamof.mofname)+f"({element})"
        mofs_ls[idx].extend(aiidamofs.mofs)

# MIL-121 binding energy comes from workflow pk = 792213, E = -32.136719775431
aiidamofs = merge_aiida_mofs_list(mofs_ls[0], mofs_ls[1])

energy_dict = {}
for aiidamof in aiidamofs:
    aiidamof.mofname = custom_uppercase(aiidamof.mofname)
    energy_dict[aiidamof.mofname] = aiidamof.ff_energy.copy()
    if aiidamof.mofname == "MIL121(Al)":
        energy_dict[aiidamof.mofname]['DFT'] = -32.136719775431
    else:
        energy_dict[aiidamof.mofname]['DFT'] = aiidamof.dft_energy

```

```python
name_ls = []
uff_ls = []
dft_ls = []
refined_ls = []
forcefield_name = ["Al_graphite","UFF"]
for name in energy_dict.keys():
    if not energy_dict[name]['DFT']:
        continue
    name_ls.append(name)
    uff_ls.append(energy_dict[name][forcefield_name[0]])
    refined_ls.append(energy_dict[name][forcefield_name[1]])
    dft_ls.append(energy_dict[name]['DFT'])

import matplotlib.pyplot as plt
fz=16
plt.scatter(range(0,len(name_ls)),uff_ls, label=f"UFF")
plt.scatter(range(0,len(name_ls)),refined_ls, label=f"Refined FF")
plt.scatter(range(0,len(name_ls)),dft_ls, label="Binding Energy (CP2K)")
plt.ylabel("Binding Energy KJ/mol",fontsize=fz)
#plt.xlabel("Cif file name",fontsize=fz)
plt.xticks(range(0,len(name_ls)), name_ls, rotation=30)  
plt.legend()
```

```python
def merge_aiida_mofs_list(aiida_mofs1, aiida_mofs2):
    mof_dict = {}
    for aiida_mof in aiida_mofs1+aiida_mofs2:
        if not aiida_mof.co_isotherms:
            continue
        if aiida_mof.mofname not in mof_dict:
            mof_dict[aiida_mof.mofname] = aiida_mof
        else:
            if mof_dict[aiida_mof.mofname].ff_energy or aiida_mof.ff_energy:
                if mof_dict[aiida_mof.mofname].ff_energy:
                    mof_dict[aiida_mof.mofname].ff_energy.update(aiida_mof.ff_energy)
                elif aiida_mof.ff_energy:
                    mof_dict[aiida_mof.mofname].ff_energy = aiida_mof.ff_energy
            if mof_dict[aiida_mof.mofname].dft_energy or aiida_mof.dft_energy:
                if not mof_dict[aiida_mof.mofname].dft_energy:
                    mof_dict[aiida_mof.mofname].dft_energy = aiida_mof.dft_energy
    return list(mof_dict.values())
```

```python
def plot_isotherm(aiidamofs):
    # global parameter
    fz = 18
    aiidamofs = [aiidamof for aiidamof in aiidamofs if aiidamof.co_isotherms]
    
    fig, axs = create_fig_axs(len(aiidamofs))
    idx = 0
    for aiidamof in aiidamofs:
        if aiidamof.co_isotherms: 
            i, j = divmod(idx, 3) # the certain way to calculate positions of figures
            for keys, values in aiidamof.co_isotherms.items():
                if keys == 'pressure':
                    continue
                ax = axs[i,j]
                ax.plot(aiidamof.co_isotherms['pressure'], aiidamof.co_isotherms[keys], marker='D', label=keys)
                ax.set_title(f"{aiidamof.mofname} at {aiidamof.temperature} K" ,fontsize=fz)
                ax.legend(fontsize=fz)
                ax.set_ylabel(f"{aiidamof.isotherm_unit}", fontsize=fz)
                ax.set_xlabel("Pressure (bar)", fontsize=fz)
                ax.tick_params(axis='both', which='major', labelsize=fz-4)
                ax.tick_params(axis='both', which='minor', labelsize=fz-8)
            idx += 1
    plt.show()
```
