## Plot pictures used in paper

```python
Figure_path = "../dataset/figures/"
```

```python
Isotherm_unit = "mol/kg" 
filename = "UFF-Al-MOFs-kg.png"
import os 
import matplotlib.pyplot as plt
import numpy as np
from aiidamof import AiidaMofs

fz = 14

aiidamofs = AiidaMofs("UFF_Al_paper.log",Isotherm_unit)
max_loading = 8
FF_name = aiidamofs[0].forcefield
# plot reference
plt.plot(np.linspace(-max_loading, max_loading, 100), np.linspace(-max_loading, max_loading, 100), 'k--')
# plot point
for aiidamof in aiidamofs:
    if aiidamof.co_isotherms:
        plt.scatter(aiidamof.co_isotherms["experiment"], aiidamof.co_isotherms[FF_name], label=aiidamof.mofname)
plt.xlim(-0.1*max_loading, max_loading)
plt.ylim(-0.1*max_loading, max_loading)
plt.xlabel(Isotherm_unit, fontsize=fz)
plt.ylabel(Isotherm_unit, fontsize=fz)
plt.title(FF_name + " For Al-MOFs", fontsize=fz)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend(fontsize=fz-4)
plt.savefig(os.path.join(Figure_path, filename))

```
