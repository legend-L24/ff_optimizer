```python
import os
import numpy as np
from isotherm import generateconfig, submit_mof, submit_mofs


# predefine a pressure list and generate the config files for each pressure
ff_data = "Al_graphite.json" 
cif_directory = "/home/yutao/project/final_cifs/"
pressure_list = [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.008, 0.01, 0.012, 0.014, 0.016, 0.018, 0.02, 0.025, 0.03, 0.035, 0.04, 0.045, 0.05, 0.055, 0.06, 0.065, 0.07] 
for temperature in [298.15, 313.15, 328.15]:
    for molecule in ["h2o"]:
        basename, _ = os.path.splitext(ff_data)
        output_name = f"{basename}_{molecule}_elias_{int(temperature)}.log"
        cif_files = [file for file in os.listdir(cif_directory) if file.endswith('.cif')]
        for cif_file in cif_files:
            cif_path = os.path.join(cif_directory, cif_file)
            generateconfig(ff_data, temperature, pressure_list, cif_path, output_name,molecule)
            submit_mof(output_name)
```

```python
import csv
def write_to_csv(aiida_mofs, filename):
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['cifname',  'ff_pressure(bar)' ,'ff_loading(mol/kg)']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for aiida_mof in aiida_mofs:
            writer.writerow({'cifname': aiida_mof.cifname, 'ff_loading(mol/kg)': list(aiida_mof.ff_loading), 'ff_pressure(bar)': list(aiida_mof.ff_pressure)})
```
